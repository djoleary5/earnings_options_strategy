```python
##DEPENDENCIES
import pandas as pd
import datetime
from datetime import timedelta, date
import yfinance as yf
import numpy as np
import yahoofinancials
from yahoofinancials import YahooFinancials
```

```python
change_df = pd.read_csv('resources/2019Q1_sp_earnings.csv', index_col=False)
change_df = change_df[['Symbol']]
```

```python
quarters = ['2019Q1', '2019Q2', '2019Q3', '2019Q4', '2020Q1']
for q in quarters:
    qtr = pd.read_csv('resources/'+q+'_sp_earnings.csv', index_col=False)
    change_df[q+' Date'] = qtr[q+' Date']
    change_df[q+' Time'] = qtr[q+' Time']
    change_df[q+' Before Open %Change'] = qtr[q+' Before Open %Change']
    change_df[q+' After Close %Change'] = qtr[q+' After Close %Change']
change_df.to_csv('resources/all_qtr_changes.csv')
change_df
```
