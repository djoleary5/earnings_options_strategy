```python
##DEPENDENCIES
import pandas as pd
import datetime
from datetime import timedelta, date
import yfinance as yf
import numpy as np
import yahoofinancials
from yahoofinancials import YahooFinancials
```

```python
quarters = ['2019Q1', '2019Q2', '2019Q3', '2019Q4', '2020Q1']
curr_qtr = '2020Q2'
```

```python
change_df = pd.read_csv('resources/2019Q1_sp_earnings.csv', index_col=False)
change_df = change_df[['Symbol']]

for q in quarters:
    qtr = pd.read_csv('resources/'+q+'_sp_earnings.csv', index_col=False)
    change_df[q+' Date'] = qtr[q+' Date']
    change_df[q+' Time'] = qtr[q+' Time']
    change_df[q+' Before Open %Change'] = qtr[q+' Before Open %Change']
    change_df[q+' After Close %Change'] = qtr[q+' After Close %Change']
change_df.to_csv('resources/all_qtr_changes.csv')
change_df
```

```python
yes_maybe_no_df = pd.read_csv('resources/2019Q1_sp_earnings.csv', index_col=False)
yes_maybe_no_df = change_df[['Symbol']]
for q in quarters:
    qtr = pd.read_csv('resources/'+q+'_sp_earnings.csv', index_col=False)
    qtr = pd.DataFrame(qtr)
    changes = []
    for row in qtr.iterrows():
        if abs(row[1][q+' Before Open %Change']) > abs(row[1][q+' After Close %Change']):
            changes.append(row[1][q+' Before Open %Change'])
        else:
            changes.append(row[1][q+' After Close %Change'])
    yes_maybe_no_df[q+' %Change'] = changes
    
yes_maybe_no_df.head()
```

```python
def yes_maybe(row):
    yes = 0
    if abs(row[1]) > 4:
        maybe = 1
    else:
        maybe = 0
    i=2
    while i<len(row):
        if abs(row[i]) > 4:
            yes = yes + 1
            maybe = maybe + 1
        i=i+1
    candidate = '3 no'
    if yes >= 3:
        candidate = '1 yes'
    elif maybe >= 3:
        candidate = '2 maybe'
    #print(row[0] + " " + str(yes) + " " + str(maybe))
    return candidate;        
```

```python
s = 0
candidates = []
while s < len(yes_maybe_no_df['Symbol']):
    row = list(yes_maybe_no_df.iloc[s])
    candidate = yes_maybe(row)
    #print(row)
    candidates.append(candidate)
    s=s+1
yes_maybe_no_df['Buy Straddle?'] = candidates
#yes_maybe_no_df = yes_maybe_no_df.sort_values(by=['Buy Straddle?', 'Symbol'])
#yes_maybe_no_df.to_csv('resources/'+q+'_sp_earnings.csv')
yes_maybe_no_df
```

```python
curr_qtr_df = pd.read_csv('resources/'+curr_qtr+'_sp_earnings.csv', index_col=False)
curr_qtr_df = pd.DataFrame(curr_qtr_df)
changes = []
success = []
for row in curr_qtr_df.iterrows():
    if abs(row[1][curr_qtr+' Before Open %Change']) > abs(row[1][curr_qtr+' After Close %Change']):
        chng = row[1][curr_qtr+' Before Open %Change']
        changes.append(chng)
    else:
        chng = row[1][curr_qtr+' After Close %Change']
        changes.append(chng)
    if abs(chng) <= 4:
        success.append('no')
    elif abs(chng) > 4:
        success.append('yes')    
    else:
        success.append('NaN')
yes_maybe_no_df[curr_qtr+' Date'] = curr_qtr_df[curr_qtr+' Date']
yes_maybe_no_df[curr_qtr+' %Change'] = changes
yes_maybe_no_df[curr_qtr+' Success'] = success
yes_maybe_no_df.to_csv('resources/'+curr_qtr+'_sp_success.csv')
yes_maybe_no_df
```

```python
yes_df = yes_maybe_no_df.loc[(yes_maybe_no_df['Buy Straddle?'] == '1 yes')]
yes_df = yes_df.sort_values(by=['Buy Straddle?', 'Symbol'])
yes_df
```
